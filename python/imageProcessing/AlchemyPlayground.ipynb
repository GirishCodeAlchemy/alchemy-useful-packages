## Resize the Image using [Pillow](https://pypi.org/project/pillow/)

```python
from PIL import Image

img = Image.open("../assets/profile.png")

# Convert the image to RGB mode
rgb_image = img.convert("RGB")

# Resize the image
resized_image = rgb_image.resize((500, 500))
```

```python
from IPython.display import display
# Display the resized image
print("Original Image Dimensions:", img.size)
display(rgb_image)
print("Resized Image Dimensions:", resized_image.size)
display(resized_image)
rgb_image.thumbnail((600, 400))
print("Thumbnail Image Dimensions:", rgb_image.size)
display(rgb_image)
```
