## Text Image Banner with [WordCloud](https://pypi.org/project/wordcloud/)

pip install wordcloud

````python

from wordcloud import WordCloud

w = WordCloud(width=1200, height=300)
i = w.generate("""Developer Python Golang Nodejs Angular,
    Frontend Backend Devops ,
    AWS GCP Azure,
    Mircoservices Terraform Jenkins Ansible Docker Kubernetes Argocd Fluxcd Helm kafka Elasticsearch,
    Kibana Datadog Splunk Prometheus""")

i
````

````python
from IPython.display import display

display(i.to_image())
````

````python
i.to_file("linkedin-banner.png")
````
